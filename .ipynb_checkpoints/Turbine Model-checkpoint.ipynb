```python
#Imports used in this project
import tensorflow.keras as kr
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.models import Model
import numpy as np
np.set_printoptions(precision=3, suppress=True)
import pandas as pd
import matplotlib.pyplot as plt

#Reads in our CSV file we got from learnonline.
powerData = pd.read_csv(
    "powerproduction.csv")

x = powerData.iloc[:,0].values
y = powerData.iloc[:,1].values

xArray = np.array(x)
yArray = np.array(y)

#Layers
krInput = kr.Input(shape=(1,))
layer1 = Dense(10, activation='sigmoid')(krInput)
layer2 = Dense(50, activation='sigmoid')(layer1)
layer3 = Dense(50, activation='sigmoid')(layer2)
layer4 = Dense(50, activation='sigmoid')(layer3)
layer5 = Dense(50, activation='sigmoid')(layer4)
layer6 = Dense(50, activation='linear')(layer5)
output = Dense(1)(layer6)

model = Model(inputs=krInput, outputs=[output])
model.compile(optimizer='adam', loss=['mean_squared_error'])

model.fit(xArray, yArray,epochs=5000, batch_size=500)
model.predict([1.001])

#Saves model so it can be used for making predictions.
model.save("Output_Model")
```
